# Ejercicio Embrague

![title](E-embP_13-71.png)

>Suponga que el sistema trabaja de manera horizontal. La fricción de las zapatas A contra el marco B hace funcional el embrague centrífugo y se transmite el torque de A hacia B. La fuerza centrífuga mantiene a las zapatas contra el marco B donde su coeficiente de fricción estático es de $\mu=0.8$. Cada zapata pesa 9.8[N] y el peso del resto de partes es despreciable. 
- Calcular la tensión inicial de los resortes (C) para que el contacto de las zapatas se haga cuando $\dot{\theta}\geq200[rpm]$
- Determinar el torque en el intervalo $100\leq\dot{\theta}\leq1500[rpm]$

# Datos 
$$\mu=0.8$$
$$mg_{zapata}=9.8[N]$$  $$m_{zapata}=\frac{9.8}{9.81}$$
![subtitle](DIAGRAMA.jpeg)
>Definimos el sistema de referecia inercial O y a su vez el sistema auxiliar radial transversal como se indica en la figura.


![subtitle](DCL.jpeg)

>Observamos las fuerzas que interactuan en los ejes radial y transversal. Y también sabemos que:
$$\vec{a}=corriolis+euler+centripeta+radial=2*\dot{r}*\dot{\theta}\vec{e_{\theta}}+r*\ddot{r}\vec{e_{\theta}}-r*\dot{\theta}^2\vec{e_{r}}+\ddot{r}\vec{e_{r}}$$
En este caso en particular $\dot{\theta}$ es cte por lo que tenemos $\ddot{\theta}=0$ y $\ddot{r}=0$
Luego tenemos: $$\sum \vec{F_{r}}=m*\vec{ar}$$ 
como: $\vec{ar}=\ddot{r}-r*\dot{\theta}^{2}*\vec{e_{r}}$ y $$\ddot{r}=0$$ 
entonces en el eje radial tenemos: $-T-N=m*(r*\dot{\theta}^2)$
En un primer instante sabemos que existe contacto por lo que la fuerza $\vec{N}$ existe pero es despreciable por lo que se considera casi nula. Y tenemos:
$$T_{o}=-m*r*\dot{\theta}^2$$
En un primer instante tenemos $r=0,3[m]$ y $\dot{\theta}=\frac{20*\pi}{3}$
$$T_{o}=\frac{9.8}{9.81}(0.3)\left(\frac{20*\pi}{3}\right)^2$$
$$\vec{T_{o}}=-131.4606 \vec{er} [N]$$
Despues tenemos:
$$N=m*r*\dot{\theta}^2-T_{o}$$
$$N=\frac{9.8}{9.81}*0.3*\dot{\theta}^2-131.4606$$
$$N=\frac{98}{327}*\dot{\theta}^2-131.4606$$

>Sabemos que la fricción es:
$$\vec{Fr}=\mu*\vec{N}=0.8*\vec{N}$$
$$Fr=0.8\left(\frac{98}{327}*\dot{\theta}^2-131.4606\right)$$
$$Fr=\frac{392}{635}*\dot{\theta}^2-105.1685$$
$$\vec{Fr}_{A/B}=-\left(\frac{392}{635}\dot{\theta}^2-105,1685\right) \vec{e\theta}$$
Sabemos que el torque se calcula con el producto cruz entre la posición y la fuerza, por lo tanto:
$$\sum\overrightarrow{\tau_o }=\overrightarrow{r}\times \overrightarrow{\mathscr{F} }$$

$$\sum\overrightarrow{\tau_o }=2(0.4)\overrightarrow{e_r}\times \left(\frac{392}{635}\dot{\theta}^{2} - 105,1685\right) \overrightarrow{\mathscr{e_\theta} })$$

$$\sum\overrightarrow{\tau_o }=2(0.4)\left(\frac{392}{635}\dot{\theta}^{2} - 105,1685\right)(\overrightarrow{e_r}\times  \overrightarrow{\mathscr{e_\theta} })$$

$$\sum\overrightarrow{\tau_o }=\left(\frac{1568}{3175}\dot{\theta}^{2} - 84,1348\right)(-\overrightarrow{k})$$

$$\sum\overrightarrow{\tau_o }=-\left(\frac{1568}{3175}\dot{\theta}^{2} - 84,1348\right)\overrightarrow{k} [Nm]$$



```python
#datos
a = 1568/3175 
b = 84.1348-105.01
#análisis
thII=np.arange((20*np.pi)/3,50*np.pi,1) #creando lista numérica
vthII = np.array(np.deg2rad(thII)) #transformado a una matriz
vsII= ((a*(vthII**2))-b)

thI=np.arange((10*np.pi)/3,(20*np.pi)/3,1) #creando lista numérica
vthI = np.array(np.deg2rad(thI)) #transformado a una matriz
vsI= (((a*(vthI**2))-b))*0

#plot velocidades
fig=go.Figure()

fig.add_trace(go.Scatter(x=thI, y=vsI, 
                         mode='lines', 
                         name='Torque 100-200'))
fig.add_trace(go.Scatter(x=thII, y=vsII, 
                         mode='lines', 
                         name='Torque 200-1500'))
 
fig.update_layout(title='Torque en funcion de la velocidad angular',
                   xaxis_title='Velocidad angular [rad/s]',
                   yaxis_title='Torque [Nm]')
fig.show()
```
